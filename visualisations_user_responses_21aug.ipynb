```python
# Read in files
control = open('text_wordcloud_control.txt', 'rb').read().decode('utf-16')
NER = open('text_wordcloud_NER.txt', 'rb').read().decode('utf-16')
POS = open('text_wordcloud_POS.txt', 'rb').read().decode('utf-16')
```

```python
control
```

```python
# Word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off");
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
# Control
wordcloud_control = WordCloud(width= 3000, height = 2000, random_state=1, collocations=False, stopwords = STOPWORDS).generate(control)
# Plot
plot_cloud(wordcloud_control)
```

```python
# NER
wordcloud_NER = WordCloud(width= 3000, height = 2000, random_state=1, collocations=False, stopwords = STOPWORDS).generate(NER)
# Plot
plot_cloud(wordcloud_NER)
```

```python
# POS
wordcloud_POS = WordCloud(width= 3000, height = 2000, random_state=1, collocations=False, stopwords = STOPWORDS).generate(POS)
# Plot
plot_cloud(wordcloud_POS)
```

Make barchart of time responses

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
x = np.arange(3)
y1 = [16.1, 6.51, 2.25]
y2 = [13.21, 3.08, 2.32]
y3 = [16.03, 7.51, 5.4]
width = 0.2
  
# Plot
plt.bar(x-0.2, y1, width)
plt.bar(x, y2, width)
plt.bar(x+0.2, y3, width)
plt.xticks(x, ['Control', 'POS', 'NER'])
plt.xlabel("Group")
plt.ylabel("Time")
plt.title("AB testing")
#plt.legend(["Control", "POS", "NER"])
plt.show()
```
