```python
# Load article
doc_2010 = open('document_2010_foreward.txt', 'rb').read().decode('ISO-8859-1')
# Process
doc = nlp(doc_2010)
# Find predicted entities
result = []
for ent in doc.ents:
    result.append((ent, ent.text, ent.start_char, ent.end_char, ent.label_))
# Save to excel
df = pandas.DataFrame(data=result)
df.to_csv("NER_2010_predicted.csv", sep=',',index=False)
#files.download("NER_2010_predicted.csv")
```

Calculate accuracy

```python
# Read in file
data = read_csv("merge_2010_true_pred_4aug.csv")
# Converting cols to list
y_true = data['y_true'].tolist()
y_pred = data['y_predicted'].tolist()
# Calculate
precision_2010 = precision_score(y_true, y_pred, average= 'macro')
recall_2010 = recall_score(y_true, y_pred, average= 'macro')
f1_score_2010 = f1_score(y_true, y_pred, average= 'macro')
accuracy_score_2010 = accuracy_score(y_true, y_pred)
# Print
print(precision_2010)
print(recall_2010)
print(f1_score_2010)
#print(accuracy_score_2010)
```

2010

```python
# Load article
doc_2010 = open('document_2010_foreward.txt', 'rb').read().decode('ISO-8859-1')
# Process
doc = nlp(doc_2010)
# Get POS tags
result_2010 = []
for token in doc:
    result_2010.append((token.text, token.pos_, token.is_stop))
# Write to df 
df = pandas.DataFrame(data=result_2010,columns=['word', 'y_predicted', 'stop_words'])
# Convert CSV
df.to_csv("POS_2010_predicted.csv", sep=',',index=False)
files.download("POS_2010_predicted.csv")
```
